# Разбираемся в неразобранном

## Сохранение и отправка на проверку

Частые ошибки при отправке задач на проверку:

1. Отправлять на проверку нужно тот код, который вы написали в пустом файле, а не консоль. Консоль - это где программа выполняется, в ней есть характерные стрелочки в начале строк. Для того, чтобы начать писать код в новом файле, нужно нажать `File` - `New file`, в открывшемся пустом окне написать свою программу и именно этот файл отправлять на проверку.
2. Расширение у ваших файлов должно быть `.py` - это две латинские буквы, сокращение от слова `Python`. НЕ кириллические буквы от слова "русский".
3. Если вы не можете загружать файлы с расширением `.py`, загружайте `.txt`. Это расширение доступно в любом текстовом редакторе, для его выбора нужно "Сохранить как" и под строкой ввода имени файла изменить тип/расширение файла.

## Ошибки при написании кода

### Если сложно - упрощай!

Если задача никак не решается, попробуйте ее упростить. Как? Пишите свой код в отдельных строках. Например вместо `print(len(word.strip()))` можно написать так:

```python
word = 'Привет! '
clean_word = word.strip()
length = len(clean_word)
print(length)
```

Это очень поможет разложить все в голове "по полочкам". Еще легче вам станет, если вы будете подписывать в комментариях, что у вас происходит в каждой строке кода:

```python
word = 'Привет! ' ## сохраняем слово в переменной word
clean_word = word.strip() ## clean_word - слово без лишних пробелов
length = len(clean_word) ## length - длина слова без пробелов
print(length) ## печатаем длину
```

Кроме того, часто можно разбить задачу на более простые шаги. Пример - как разбиты на подзадачи наши задачи с общим номером. Если бы задачу выдавать одним условием, оно бы звучало так:

    Пользователь вводит свои фамилию, имя и отчество - каждое в отдельной строке "анкеты". Программа должна напечатать в одной строке разделенными одним пробелом фамилию и инициалы, избавившись от лишних пробелов, заменив все "ё" на "е" и убедившись в корректности регистра: фамилия должна быть написана строчными буквами, но начинаться с заглавной, инициалы должны быть заглавными буквами, после инициалов должны стоять точки.

Над такой задачей пришлось бы некоторое время подумать прежде чем вообще приступать к решению. Но эту часть работы за вас уже выполнили, чтобы вы могли сосредоточиться на отработке написания кода. Вы можете аналогичным образом разбивать для себя на отдельные шаги написание и более простых программ.

### input()

В аргумент `input()` передается подсказка пользователю, руководство к действию. Если это сложная формулировка, можно просто считать, что первое слово в скобках всегда должно быть "Введите". Например `input('Введите число: ')`.

Если вы хотите, чтобы ваш ввод не приклеивался к подсказке пользователю, и все выглядело красиво, в конце подсказки ставьте пробел. Если вы запустили программу и на "Введите число" ввели ` 5`, питон, как правило, выдаст ошибку, потому что он не знает, как сделать числом пробел.

### Кавычки

**Все, что написано в кавычках - это те значки, которые написаны в кавычках. Все, что написано без кавычек - это переменные, функции или операторы: то, что питон должен выполнить. Будьте внимательны.** Если вы напишете название переменной в кавычках, программа вам просто напечатает название этой переменной, а не то, что вы в ней сохранили. И наоборот, если вы написали без кавычек что-то, что питон не должен интерпретировать как переменную, он выдаст ошибку, поскольку вы ничего не записывали в переменную с названием, например, `привет`.

## Дебаггинг

Дебаггинг - это процесс поиска ошибок в коде. В реальной жизни, где программы состоят не из десяти строк, а из сотен и тысяч, кода без ошибок не бывает. Невозможно нигде не забыть запятую, ни разу не опечататься и не запутаться в логике.

У ошибок бывает несколько причин:

* `синтаксические ошибки` возникают из-за несоблюдения синтаксиса: где-то забыта скобка, точка, запятая, отступ или что-то еще, что не позволяет интерпретатору корректно прочитать код. Простые синтаксические ошибки IDLE ловит еще до запуска программы и выдает всплывающее окно `Syntax error`;
* `runtime`: эти ошибки не подсвечиваются в среде разработки и могут не проявиться при запуске, но в процессе работы с приложением они его «положат»;
* `ошибки логики` не помешают программе отработать, но результат работы будет не такой, как мы ожидаем. Такие ошибки найти сложнее всего, поскольку их среда разработки не подсветит и никак на них не укажет: с ее точки зрения, раз программа работает, значит, все в порядке.

Как находить свои ошибки?

### Запуск кода

Если в вашем коде есть синтаксические или runtime ошибки, многие из них вы можете легко найти, просто запустив свой код. Не пренебрегайте этим шагом. Будет очень обидно, если вы на контрольной сдадите сложную задачу, которая будет полностью правильно решена, но код окажется нерабочим из-за пропущенной скобки. Такая задача может быть зачтена на 0 баллов. Не оказывайтесь в такой ситуации.

### Ошибка есть, а в чем она заключается?

Допустим, у нас есть код, который выдает ошибку:

```python
my_variable = 'Привет!'

print(my_variabe)
```

Среда разработки четко определила не только где возникла ошибка, но и в чем она заключается, надо только уметь это прочитать. Давайте посмотрим внимательно:

`Traceback (most recent call last)` - неотъемлемая и неизменная строка в любой ошибке. Она нас не интересует.

`Cell In[1], line 3` - адрес ошибки. В Colab это, как и у меня, ячейка In с каким-то номером, у вас - адрес расположения файла с кодом на компьютере. После адреса ячейки или файла указывается номер строки, в которой произошла ошибка. Указание строки очень помогает в нахождении ошибки - особенно если в вашем коде много строк или есть несколько очень похожих.

    В IDLE по умолчанию выключено отображение номеров строк, но можно его включить в параметрах среды разработки.

`----> 3 print(my_variabe)` - цитирование строки, в которой произошла ошибка.

`NameError: name 'my_variabe' is not defined` - тип и текст ошибки. `NameError` - это тип, здесь - ошибка имени. `name 'my_variabe' is not defined` - текст: имя 'my_variabe' не определено. Оно не определено, потому что мы нигде не задавали переменную my_variabe. Кстати, создание переменных и функций еще называют определением этих переменных и функций. Отсюда и слово "не определено".

Из этой ошибки мы можем сделать вывод, что программа пытается найти переменную `my_variabe`, чтобы прочитать и напечатать ее значение, но не может это сделать. Мы знаем, что в этой строке должно печататься значение нашей переменной. Что могло пойти не так? Видимо, когда мы переменную создавали, мы дали ей не то имя, которое пытается найти программа. Если посмотреть внимательно - действительно, при вызове переменной для печати у нас произошла опечатка.

Внимательно читайте текст ошибок: в нем часто будет содержаться ответ на ваш вопрос.
