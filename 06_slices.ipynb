# Срезы

Срезы немного похожи на индексы и немного похожи на `range()` и позволяют получать целые куски строки, причем не обязательно идущие подряд.

Пишется это так: [`начало`:`конец`]. Разберемся по порядку.

Начало всегда включительно, конец всегда не включительно.

```python
line
```

```python
line[2:5]
```

```python
line[0:4]
```

```python
line[:4] # можно пропустить начало, тогда началом будет начало строки
```

```python
line[5:] # можно пропустить конец, тогда концом будет конец строки
```

```python
line[:] # можно сделать так, хотя в этом нет никакого смысла
```

Третьим аргументом срезу можно задать шаг. Например, если шаг 2, то мы получим каждый второй символ строки.
[`начало`:`конец`:`шаг`]

```python
nums = '0123456789'
```

```python
nums[1:8:2]
```

```python
nums[::3] # почему важно поставить два двоеточия?
```

```python
nums[:8:2]
```

Помним, что индексы могут быть отрицательными? Так вот, аргументы срезов тоже могут.

```python
nums[-7:-2]
```

```python
nums[-7:7]
```

```python
nums[-2:-7] # почему получилась пустая строка?
```

```python
nums[8:1:-1]
```

```python
nums[1:8:-1]
```

```python
nums[::-1]
```

```python
nums[-2::-2]
```
