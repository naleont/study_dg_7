## Методы

Для строк существует большое количество методов. Метод - это что-то, очень похожее на функцию, но отличается тем, что может применяться только к определенному типу данных.

Например `print()` работает на любых данных. А вот `.upper()` на числах не сработает. Метод приводит строку к верхнему регистру. Для чисел он не имеет смысла, потому что у чисел нет регистра.

Обратите внимание на синтаксис: методы, в отличие от функций, пишутся после данных, к которым они будут применены, через точку. Скобки, как и у функций, обязательны, даже если они пустые.

```python
line.upper()
```

Метод `.lower()`, наоборот, приводит строку к нижнему регистру.

```python
'EXPILLIARMUS'.lower()
```

Метод `.replace()` принимает два аргумента. Он находит в строке подстроку (символ или последовательность символов), переданную первым аргументом, и заменяет на подстроку, переданную вторым аргументом.

```python
doll = 'матрёшка'

doll.replace('ё', 'е')
```

Мы можем передать этому методу пустую строку в качестве второго аргумента. Что получится?

```python
kettle = 'скот'

kettle.replace('с', '')
```

Обратите внимание на особенности срабатывания метода в случае, если подстрока найдена больше одного раза: метод обрабатывает все вхождения подстроки, то есть все разы, когда подстрока была найдена.

```python
'уруруруруру'.replace('у', 'а')
```

## Индексы

Строка - это символы, расположенные в определенном порядке. Значит, у каждого символа есть порядковый номер. Такой номер называют **индексом**.

**Счёт в Python начинается с 0.**

```python
line[0]
```

```python
'witch'[0]
```

```python
line[1]
```

```python
len(line)
```

А как получить последний символ строки?

```python
line[8]
```

```python
line
```

```python
line[7]
```

```python
w = 'goodbye'
```

```python
len(w)
```

```python
w[6]
```

Но программа, которую мы пишем, должна быть универсальной. Странно писать отдельную программу для всех вариантов длины слова, которые могут встретиться. Можем ли мы находить последний символ строки независимо от ее длины?

```python
line[len(line) - 1]
```

```python
w[len(w) - 1]
```

Индексы могут быть отрицательными. Посмотрите, как они работают.

```python
line[-1]
```

```python
line[-2]
```

```python
line[-len(line)]
```

Есть метод для поиска подстроки в строке. Посмотрите, работает ли метод со всеми вхождениями?

```python
line.find('i')
```

В слове quiddich искомая буква встречается 2 раза. Метод вернул нам только первое вхождение.

Есть метод, который выполняет ту же задачу, но начинает поиск с конца строки.

```python
line.rfind('i')
```

А .rfind() нашел первую i с конца.

Еще один полезный метод - `.strip()`. Он удаляет ненужные символы из начала и из конца строки (в том числе если в начале или конце строки стоит несколько подобных символов подряд). Если аргументы не переданы (скобки оставлены пустыми), удалятся все пробельные символы. Это пробелы, отступы и прочие подобные символы. Если в аргумент передана подстрока, метод найдет в начале и конце нашей строки все указанные в подстроке символы и удалит их. При этом если символ из числа тех, которые нужно удалить, найден не в начале и не в конце строки, его метод не затронет. Посмотрите, как работает этот метод.

```python
line = '    hello '

line.strip()
```

```python
line = '    Hi, how are you?        '
line.strip()
```

```python
line = '?.,!punc,.tuation!,'
line.strip('!?,.')
```

### Задача 1.1

Пользователь вводит свои фамилию, имя и отчество. Программа должна напечатать ФИО в одну строку. Между фамилией, именем и отчеством должны стоять пробелы. У задачи есть как минимум 3 способа решения, подойдет любой из них.

### Задача 1.2

Пользователь вводит свои фамилию, имя и отчество. Очень часто при заполнении данных пользователи вводят лишние пробелы до или после данных. Программа должна удалить эти лишние пробелы, а затем вывести ФИО, как в предыдущей задаче.

Можно подумать, что задача решается двумя способами, но только один из них позволит корректно обрабатывать случаи, например, с двойными именами. Имя 'Светлана Полина' при удалении пробелов не должно склеиться, пробел между двумя именами должен остаться на месте.

### Задача 1.3

Будем считать, что принимаем данные для печати документов, в которых не допускается использование буквы `ё`. К решению предыдущей задачи нужно добавить замену всех `ё` на `е`.

### Задача 1.4

Между фамилией и обоимии инициалами должны стоять пробелы. Не забывайте, что данные перед обработкой нужно почистить от пробелов.

### Задача 1.5

Задача аналогична предыдущей, только теперь нужно вывести фамилию и инициалы в верхнем регистре. Например ИВАНОВ И. И.

### Задача 1.6

Усилим нашу очистку данных. Теперь будем не только удалять лишние пробелы, но и приводить в порядок регистр. Пользователь может вводить данные как угодно: всеми большими буквами, всеми маленькими - хоть большими и маленькими через одну. Нужно, чтобы в итоге получились фамилия с первой заглавной буквы и всеми остальными строчными и инициалы заглавными буквами: Иванов И. И.
