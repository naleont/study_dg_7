## Логические операторы

````python
# and позволяет проверять, выполняются ли оба (или больше) условий
a = int(input())
if a % 3 == 0 and a % 2 == 0:
  print('кратно 6')
````

````python
# or позволяет проверять, выполняется ли хотя бы одно из условий
a = int(input())
if a % 3 == 0 or a % 2 == 0:
  print('кратно 2 или 3')
````

````python
# not отрицает результат выражения, следующего за ним
a = int(input())
if not a % 3 == 0:
    print('не кратно 3')
````

```python
# условий, объединенных через and, or может быть сколько угодно
if a % 3 == 0 and a % 2 == 0 or a % 5 == 0:
    
# когда важен порядок проверки условий, используйте скобки
if a % 3 == 0 or (a % 2 == 0 and a % 5 == 0):
```

# Приведение типов данных

`bool` - обозначение булева типа данных. Функция `bool()` приводит любой тип данных к логическому, если это возможно.

````python
bool(1)
````

````python
bool(0)
````

````python
bool(6457)
````

````python
bool(-62)
````

````python
n = 15

if n:
  print('true')
else:
  print('false')
````

````python
n = 0

if n:
  print('true')
else:
  print('false')
````

````python
bool('hello')
````

````python
bool('goodbye')
````

````python
bool('')
````

````python
name = input('Введите имя ')

if not name:
  print('Вы ничего не ввели')
else:
  print('Ваше имя ' + name)
````

````python
name = input('Введите имя ')

if name == '':
  print('Вы ничего не ввели')
else:
  print('Ваше имя ' + name)
````

Можно обобщить правила приведения данных к булевому типу следующим образом: наличие данных всегда выдаст True, а их отсутствие - False:

`0` -> `False`\
`''` (пустая строка) -> `False`\
`[]` (пустой список) -> `False`\
`{}` (пустой словарь) -> `False`\
`None` (тип данных "ничто") -> `False`
